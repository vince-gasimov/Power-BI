```python
import pandas as pd
```

```python
# Verilerinizi yükleyin
data = pd.read_csv('job_postings - Kopya.csv')

# 'Job Skills' sütununu ele alıp her bir beceriyi yeni bir satıra dökmek için bir fonksiyon yazalım
def explode_list(df, column_to_explode, keep_columns):
    # Sütunu liste olarak ayır
    df = df.dropna(subset=[column_to_explode])
    df[column_to_explode] = df[column_to_explode].apply(eval)

    # Liste elemanlarını satırlara ayır
    s = df.apply(lambda x: pd.Series(x[column_to_explode]), axis=1).stack().reset_index(level=1, drop=True)
    s.name = column_to_explode

    # Diğer sütunları koruyarak yeni DataFrame oluştur
    return df.drop(column_to_explode, axis=1).join(s)

# Fonksiyonu kullanarak 'Job Skills' sütununu ayır
exploded_skills_df = explode_list(data, 'Job Skills', ['Job Posting ID'])

exploded_skills_df.head()
```

```python
df = exploded_skills_df[["Job Posting ID", "Job Skills"]]
```

```python
# CSV olarak kaydet
df.to_csv('exploded_job_skills.csv', index=False)
```
